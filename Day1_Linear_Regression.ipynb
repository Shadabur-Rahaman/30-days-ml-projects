```python
import numpy as np
import matplotlib.pyplot as plt

# Generate synthetic data: y = 3x + 4 + noise
np.random.seed(42)
X = 2 * np.random.rand(100, 1)
y = 3 * X + 4 + np.random.randn(100, 1)

# Initialize parameters
theta = np.random.randn(2, 1)  # weights (slope and intercept)
learning_rate = 0.1
n_iterations = 1000
m = len(X)

# Add x0 = 1 to each instance (for intercept)
X_b = np.c_[np.ones((m, 1)), X]

# Gradient Descent loop
for iteration in range(n_iterations):
    gradients = 2/m * X_b.T.dot(X_b.dot(theta) - y)
    theta = theta - learning_rate * gradients

print("Learned parameters (intercept and slope):")
print(theta)

# Plot the data and the linear fit
plt.scatter(X, y, label="Data points")
plt.plot(X, X_b.dot(theta), color='red', label="Fitted line")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.title("Linear Regression Fit")
plt.show()
```
